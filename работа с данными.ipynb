```python
VKing=1000
VQueen=100
VBishop=40
VKnight=45
VRook=60
VPawn=20
class Color(object):
    EMPTY=0
    BLACK=1
    WHITE=2
class HowCasting(object):
    NONE=1
    LEFT=2
    RIGHT=3
    ALL=6
```

```python
def FEN_to_vector(pos):
    fen=pos['fen']
    res=pos['res']
    print(fen)
    pos, move, casting, pawn, _,_= fen.split(' ')
    vector=[0]*69
    #matrix=np.zeros(69, dtype=int)
    i=0
    pos_lines=pos.split("/")
    def get_value(c):
        if(c=='p'):
            return VPawn
        if(c=='n'):
            return VKnight
        if(c=='b'):
            return VBishop
        if(c=='r'):
            return VRook
        if(c=='q'):
            return VQueen
        if(c=='k'):
            return VKing
    for pos_line in pos_lines:
        for cel in pos_line:
            if(cel.isdigit()):
                i+=int(cel)
            else:
                vector[i]=(2*cel.isupper()-1)*get_value(cel.lower())
                i+=1
    if(move=='b'):
        vector[64]=Color.BLACK
    else:
        vector[64]=Color.WHITE
    vector[65]=HowCasting.NONE
    vector[66]=HowCasting.NONE
    for cel in casting:
        if(cel.lower()=='q'):
            vector[65+cel.islower()]*=HowCasting.LEFT
        elif(cel.lower()=='k'):
            vector[65+cel.islower()]*=HowCasting.RIGHT
    if(len(pawn)==2):
        if(pawn[0]=='a'):
            ver_line=0
        elif(pawn[0]=='b'):
            ver_line=1
        elif(pawn[0]=='c'):
            ver_line=2
        elif(pawn[0]=='d'):
            ver_line=3
        elif(pawn[0]=='e'):
            ver_line=4
        elif(pawn[0]=='f'):
            ver_line=5
        elif(pawn[0]=='g'):
            ver_line=6
        elif(pawn[0]=='h'):
            ver_line=7
        vector[67]=ver_line*8+int(pawn[1])-1
    else:
        vector[67]=0
    vector[68]=res
    return vector
```
